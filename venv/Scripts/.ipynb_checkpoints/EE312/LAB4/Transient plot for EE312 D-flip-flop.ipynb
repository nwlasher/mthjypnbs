```python
import numpy as np
import matplotlib as mpl
import matplotlib . pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings ("ignore")
# Define filename
filename = 'D Flip Flop CLK D and Q.csv'
time , CLK , D_in , Q_out = np.loadtxt (filename, skiprows = 1, delimiter = ',', unpack = True)
# Plot the signals
plt.figure(figsize = (6, 6))
plt.subplot(311)
plt.title('D Flip Flop')
plt.plot(time, CLK)
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Clock')
plt.subplot(312)
plt.plot(time, D_in )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('D Input')
plt.subplot(313)
plt.plot(time, Q_out )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Q Output')
plt.xlabel('Time[s]')
plt.savefig('D Flip Flop CLK D and Q.png')
plt.show()
```
