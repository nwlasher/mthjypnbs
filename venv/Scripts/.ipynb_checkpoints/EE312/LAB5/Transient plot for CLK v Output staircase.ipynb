```python
import numpy as np
import matplotlib as mpl
import matplotlib . pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings ("ignore")
# Define filename
filename = 'DAC_CLK_vs_Output_ladder_data.csv'
time , CLK , Q_out = np.loadtxt (filename, skiprows = 1, delimiter = ',', unpack = True)
# Plot the signals
plt.figure(figsize = (6, 6))

plt.subplot(311)
plt.title('Clock input signal and Output Staircase signal vs Time')
plt.plot(time, CLK)
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Clock(Volts)')

plt.subplot(312)
plt.plot(time, Q_out )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Output Staircase(Volts)')

plt.xlabel('Time(Seconds)')
plt.savefig('CLK and Stair v Time.png')
plt.show()
```
