```python
import numpy as np
import matplotlib as mpl
import matplotlib . pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings ("ignore")
# Define filename
filename = 'DAC_CLK_BITS_and_Output_ladder_data.csv'
time , CLK , S_0 , S_1, S_2, S_3, Q_out = np.loadtxt (filename, skiprows = 1, delimiter = ',', unpack = True)
# Plot the signals
plt.figure(figsize = (8, 12))

plt.subplot(7,1,1)
plt.title('Clock input, Bit positions, and Output Staircase vs Time')
plt.plot(time, CLK)
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Clock Signal (Volts)')

plt.subplot(7,1,2)
plt.plot(time, S_0 )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Bit 0(Volts)')

plt.subplot(7,1,3)
plt.plot(time, S_1 )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Bit 1(Volts)')

plt.subplot(7,1,4)
plt.plot(time, S_2 )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Bit 2(Volts)')

plt.subplot(7,1,5)
plt.plot(time, S_3 )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Bit 3(Volts)')

plt.subplot(7,1,6)
plt.plot(time, Q_out )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Output Staircase(Volts)')

plt.xlabel('Time(Seconds)')
plt.savefig('DAC_CLK_BITS_Output_ladder_v_TIME_data.png')
plt.show()
```
