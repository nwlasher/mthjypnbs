```python
import numpy as np
import matplotlib as mpl
import matplotlib . pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings ("ignore")
# Define filename
filename = '200KHz_AnalogIN_DigitalOUT_AnalogOUT_data.csv'
time , A_in , D_out , A_out = np.loadtxt (filename, skiprows = 1, delimiter = ',', unpack = True)
# Plot the signals
plt.figure(figsize = (8, 8))
plt.subplot(311)
plt.title('Clock input, Digital output, and Analog output vs Time')
plt.plot(time, A_in)
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Analog input(Volts)')
plt.subplot(312)
plt.plot(time, D_out )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Digital output(Volts)')
plt.subplot(313)
plt.plot(time, A_out )
plt.axvline(color = 'black')
plt.axhline(color = 'black')
plt.ylabel('Analog output')
plt.xlabel('Time(Seconds)')
plt.savefig('200KHz_AnalogIN_DigitalOUT_AnalogOUT_data.png')
plt.show()
```
